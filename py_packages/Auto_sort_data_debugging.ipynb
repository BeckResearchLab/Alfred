```python
def find_directory():
    '''
    Assuming your python file is in the directory containing KCM data files,
    returns a path to that directory with an additional
    forward slash for future concatenation processes.
    '''
    path = pathlib.Path().absolute()
    directory = str(path) + '/'
    return directory
```

```python
def grab_csv(directory):
    '''
    Returns a list of all csv files in an existing directory.
    Ignoring the lone xlsx file for now, as reader fails to read this file.
    '''
    list_of_csv = []
    for filename in listdir(directory):
        if (filename.endswith('.csv')):  # or filename.endswith('.xlsx')):
            list_of_csv.append(filename)
        else:
            pass
    return list_of_csv
```

```python
import zipfile
import os
import tempfile

def extract_zip(zip_file_path, extract_to):
    """
    Extracts the contents of a zip file to a specified directory.
    
    Args:
    - zip_file_path: Path to the zip file.
    - extract_to: Directory to extract the contents to.
    
    Returns:
    - List of extracted file paths.
    """
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
        extracted_files = zip_ref.namelist()  # Get the list of extracted file paths
    return extracted_files

def grab_csv_from_zip(zip_file_path):
    """
    Extracts the contents of a zip file and returns a list of all CSV files.
    
    Args:
    - zip_file_path: Path to the zip file.
    
    Returns:
    - List of CSV file paths.
    """
    temp_dir = tempfile.mkdtemp()  # Create a temporary directory to extract files
    extracted_files = extract_zip(zip_file_path, temp_dir)
    
    # Filter out CSV files
    csv_files = [os.path.join(temp_dir, file) for file in extracted_files if file.endswith('.csv')]
    
    return csv_files

# Example usage:
zip_file_path = '/home/clgould99/TheBattmobile/py_packages/Raw Data/KCM-Raw-Data.zip'
csv_files_list = grab_csv_from_zip(zip_file_path)
print("CSV Files in Zip:", csv_files_list)
```
