```python
import matplotlib.pyplot as plt
new_df=build_data_vis.build_module_df(directory, 'bus_100/', 1)
df_bus_dates = build_data_vis.sort_bus_by_date(directory, 'bus_100/')
no_of_files=df_bus_dates.shape[0]
sep_dfs=[]
for i in range (no_of_files):
    start=12*i
    end=12*(i+1)
    sep_df=new_df.iloc[start:end]
    sep_dfs.append(sep_df)


voltage_values = [2 + i * 0.2 for i in range(11)]
# Iterate through each row and plot a histogram
for j in range (no_of_files): 
    sep_dfs[j]=sep_dfs[j].iloc[:,1:-1]
    for i in range(len(sep_dfs[j])):     
        row = sep_dfs[j].iloc[i]
        plt.bar(voltage_values, height=row, width=0.2)
        plt.xlabel('Voltage')
        plt.ylabel('Time')
        plt.title(f'Voltage Distribution for Module{i+1} for File {j+1}')
        plt.show()
```

```python
import matplotlib.pyplot as plt

new_df=build_data_vis.build_module_temp(directory, 'bus_110/', 1) # get temperature distribution vs time
df_bus_dates = build_data_vis.sort_bus_by_date(directory, 'bus_110/')
no_of_files=df_bus_dates.shape[0]

#Separate the total dataframe into file by file dataframe
sep_dfs=[]
for i in range (no_of_files):
    start=2*i
    end=2*(i+1)
    sep_df=new_df.iloc[start:end]
    sep_dfs.append(sep_df)


temp_values = [-5 + i * 5 for i in range(16)]

fig, ax = plt.subplots(no_of_files, 2, figsize=(12, 12))
fig.subplots_adjust(hspace=0.3)
# Iterate through each row and plot a histogram
for j in range (no_of_files): 
    sep_dfs[j]=sep_dfs[j].iloc[:,8:-3]
    for i in range(len(sep_dfs[j])):     
        row = sep_dfs[j].iloc[i]
        ax[j,i].bar(temp_values, height=row, width=0.5)
        ax[j,i].set_xlabel('Temperature')
        ax[j,i].set_ylabel('Time')
        ax[j,i].set_title(f'Temperature vs Time for Sensor {i+1} for File {j+1}')
plt.show()
```

```python
#time of failure
mtf={}
df_bus_dates = build_data_vis.sort_bus_by_date(directory, 'bus_1')
no_of_files=df_bus_dates.shape[0]
print (no_of_files)
for i in range (1,17):
    mtf.update({f'module no{i}':[]})
print(mtf)

for module in range(1, 17):
    module_df = build_data_vis.build_module_average_df(directory, 'bus_1', module)
    
    for i in range (no_of_files-1):
        
        if (module_df.iloc[i,12]>module_df.iloc[i+1,12]):
            mtf[f'module no{module}'].append(round(module_df.iloc[i,12]))
        
        elif (module_df.iloc[i,12]<module_df.iloc[i+1,12]):
            mtf_list = mtf[f'module no{module}']
            
            if len(mtf_list) == 0 or len(mtf_list)==1:
                mtf_list.append(round(module_df.iloc[i+1, 12]))
                
                if len(mtf_list)==2 and mtf_list[1]>mtf_list[0]:               
                        mtf_list = [mtf_list[1]]
            else:
                mtf_list[-1] = round(module_df.iloc[i+1, 12])
            mtf[f'module no{module}'] = mtf_list

#print(mtf)
mtf_df=pd.DataFrame.from_dict(mtf,orient = 'index')
print(mtf_df)
```
